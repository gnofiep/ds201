```python
import pandas as pd
from bokeh.io import output_notebook, show, output_file
from bokeh.plotting import figure
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Lemonade_Lab10.csv")
df.head(5)
```

```python
p = figure(title="Lemon and Orange Sales by Temperature")
```

```python
p.circle(x=df['Temperature'], y=df["Lemon"], size=10, alpha=0.5, color="green", legend = "Lemon")
```

```python
p.triangle(x=df['Temperature'], y=df["Orange"], size=10, alpha=0.5, color="orange", legend = "Orange")
p.legend.location="top_left"
show(p)
```

```python
# 5)
# The sale for lemon increases relative to temperature, 
# while the orange remain steady relative to the temperature.
# Most of the highest sales are in the range of 80 to 84 F and 
# there are many days that are in high temperature
```

```python
grouped = df.groupby("Temperature")
lem = grouped['Lemon']
avg = lem.mean()
std = lem.std()
Tem = list(grouped.groups.keys())
Park = df[df["Location"]=="Park"]
Beach = df[df["Location"]=="Beach"]
```

```python
v=figure(title="Temperature Vs Lemon Sales (Park and Beach)")
```

```python
v.vbar(x=Tem, bottom=avg-std, top= avg+std, width=0.8, fill_alpha=0.2, line_color = None,legend= "Lemon Revenue 1 stddev")
v.circle(x=Beach["Temperature"],y=Beach["Lemon"], alpha=0.5, size=10,color='red',legend= "Sales in Beach")
v.triangle(x=Park["Temperature"],y=Beach["Lemon"], alpha=0.5, size=10,color='blue',legend="Sales in Park")
v.legend.location="top_left"
show(v)
```

```python
# 7)
# From the figure, we can deduce that there are sale only in the Park but not in Beach when the temperature are low,
# in range of 70 to 72 F. Whereas for the Beach, there are sale only in the Beach but no slae in the Park
# when the temperature are in range of 74 to 76 F
# When the temperature increase, the sales increases for both Park and Beach
# There are more days with higher temperature deduced form the figure, 
# which shown that there are more number of sales in the hot days.
```
